```python
import PyPDF2

# Open the PDF file in read-binary mode
with open("sample1.pdf", 'rb') as pdf_file:
    # Create a PDF reader object
    pdf_reader = PyPDF2.PdfReader(pdf_file)

    # Print the number of pages in the PDF file
    print(f"Number of pages: {len(pdf_reader.pages)}")

    # Get the first page
    page = pdf_reader.pages[0]

    # Extract text from the first page
    text = page.extract_text()
    print("Text from page 1:")
    print(text)
```

```python
import docx
from docx.shared import Inches

# Create an instance of a Word document
doc = docx.Document()

# Add a main heading (level 0)
doc.add_heading('Heading for the document', level=0)

# Add a paragraph and store the object in a variable
doc_para = doc.add_paragraph('Your paragraph goes here, ')

# Add styled runs (bold and italic)
doc_para.add_run('hey there, bold here').bold = True
doc_para.add_run(', and ')
doc_para.add_run('these words are italic').italic = True

# Add a page break
doc.add_page_break()

# Add a level 2 heading
doc.add_heading('Heading level 2', level=2)

# Add an image (with optional width)
try:
    doc.add_picture("index.jpg", width=Inches(4))
except Exception as e:
    print(f"Error adding picture: {e}")

# Save the document
doc.save('new_doc.docx')
```
